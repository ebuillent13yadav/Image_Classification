```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
```

```python
data_train_path = '/content/drive/MyDrive/Datasets/Fruits_Vegetables/train'
data_test_path = '/content/drive/MyDrive/Datasets/Fruits_Vegetables/test'
data_val_path = '/content/drive/MyDrive/Datasets/Fruits_Vegetables/validation'
```

```python
data_train = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle=True,
    image_size=(image_width, image_height),
    batch_size=32,
    validation_split=False
)
```
